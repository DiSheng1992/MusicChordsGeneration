```python
import pandas as pd
import numpy as np
import random as rand

df = pd.read_csv('chord.2ndL.csv')
```

```python
chordvec = list(df[df['L1'] == 1]['child_path'])
for i in range(28):
    string = chordvec[i]
    chordvec[i] = string.split(',')[0]

    
indexlist = list(df[df['L1'] == 1].index)
indexlist.append(759)

def generate_chord_pattern(chord: str, length: int)->list:
    pattern = []
    for i in range(length):
        pattern.append(chord)
        j = chordvec.index(chord)
        weights = np.array(df['probability'])[indexlist[j]: indexlist[j+1]]
        choicevec = list(df['chord_ID'])[indexlist[j]: indexlist[j+1]]
        chord = rand.choices(choicevec, weights, k=1)[0]
        while chord not in chordvec:
            chord = rand.choices(choicevec, weights, k=1)[0]
    return pattern

chord_dict = {'1': ('Cmaj', ['C2','C4','E4','G4','C5']),
              '4': ('Fmaj', ['F2','F3','A3','C4','F4']),
              '5': ('Gmaj', ['G2','G3','B3','D4','G4']),
              '6': ('Am', ['A2','A3','C4','E4','A4']),
              '2': ('Dm', ['D2','D4','F4','A4','D5']),
              '3': ('Em', ['E2','E4','G4','B4','E5']),
              '27': ('Dm7',['D2','D4','F4','A4','C5']),
              '47': ('Fmaj7',['F2','F3','A3','C4','E4']),
              '16': ('C6',['C2','C4','E4','G4','B4']),
              '67': ('Am7',['A2','A3','C4','E4','G4']),
              '56': ('G6', ['G2','G3','B3','D4','E4']),
              '5/6': ('Am9', ['A2','E3','G3','B3','D4']),
              'b7': ('Bdim7', ['B2','B3','D4','F4','G#4']),
              '164': ('C6', ['C2','C4','E4','G4','A4']),
              '57': ('G7',['G2','G3','B3','D3','F3']),
              '37': ('Em7', ['E2','E3','G3','B3','D3']),
              '5/5': ('D7', ['D2','D4','F#4','A4','C4']),
              '464': ('F6', ['F2','F3','A#4','C4','F4']),
              '17': ('C7', ['C2','C4','E4','G4','A#4']),
              '664': ('Adim', ['D2','A3','C4','E4','F#4']),
              '564': ('C9', ['F2','C4','E4','G4','D5']),
              '46': ('F6', ['F2','F3','A3','C4','D4'] ),
              '57/6': ('G7b9',['E2','G3','D4','F#4','B4']),
              '642': ('Asus2',['A2','B3','E4','A4','D5']),
              '36': ('Em6',['E2','D4','E4','G4','B4']),
              '66': ('A6',['D3','A3','C4','E4','F#4']),
              'b6': ('Amaj', ['A2','A3','C#4','E4','A4']),
              'b3': ('Emaj', ['E2','E3','G#3','B3','E4'] )
}

def num_to_chord(pattern):
    chordpattern =[]
    for chord in pattern:
        chordpattern.append(chord_dict[chord])
    return chordpattern
```
